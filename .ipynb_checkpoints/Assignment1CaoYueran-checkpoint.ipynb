The code below reads in the given CSV files named 'Assignment.csv', 'GradedTask.csv', 'Test.csv', 'Exam.csv' into the variable `rawdata`.

```python
import numpy as np
assignment_data = np.genfromtxt('Assignment.csv',delimiter = ',',dtype=str)
gradedtask_data = np.genfromtxt('GradedTask.csv',delimiter = ',',dtype=str)
test_data = np.genfromtxt('Test.csv',delimiter = ',',dtype=str)
exam_data = np.genfromtxt('Exam.csv',delimiter = ',',dtype=str)
rawdata = (assignment_data,gradedtask_data,test_data,exam_data)
rawdata
```

a) Write a function `averageMarks(data)` which will compute the average marks of each assessment for each student, converted to 100%. You should return a np array of shape (300,4), where each row contains the score of each student. Each column will be respective assessment of each student (Assignment, Graded, Test, Exam).

Sample output for first 3 and last 3 students:

<pre>array([[68.        , 83.        , 44.44444444, 85.        ],
       [64.        , 73.        , 82.22222222, 59.        ],
       [36.        , 74.        , 73.33333333, 56.        ],
       ...,
       [56.        , 64.        , 48.88888889, 81.        ],
       [64.        , 73.        , 57.77777778, 51.        ],
       [88.        , 71.        , 51.11111111, 63.        ]])</pre>

```python
def averageMarks(data):
    res = np.reshape(np.mean(data[0][2:,1:].astype(int)/data[0][1,1:].astype(int), axis=1),(-1,1))
    for x in range(1, len(data)):
        res = np.hstack((res, np.reshape(np.mean(data[x][2:,1:].astype(int)/data[x][1,1:].astype(int), axis=1),(-1,1))))
    return res
a = averageMarks(rawdata)
print(a)
```

<pre>array([[60.72, 85.  , 70.43],
       [75.81, 59.  , 69.09],
       [66.07, 56.  , 62.04],
       [67.31, 53.  , 61.59],
       [73.42, 51.  , 64.45]])</pre>

```python
# Your Solution
def computemarks(data):
    a = np.hstack((np.reshape(np.sum(data[:,:3]*np.array([.2,.3,.5]),axis=1),(-1,1)), np.reshape(data[:,3], (-1,1))))
    b = np.hstack((a, np.reshape(a[:,0]*.6+a[:,1]*.4, (-1,1))))
    return np.round(b*100, 2)
b = computemarks(a)
print(b)
```

You should return a 1D array containing the corresponding grade of each student.

Sample output for first 5 students:
<pre>array([['A'],
       ['B'],
       ['B'],
       ['B'],
       ['B']], dtype='&lt;U2')</pre>

```python
# Your Solution
def grades(data):
    def g(n):
        k = n[2]
        if k >= 80:
            return 'A+'
        elif k >= 70:
            return 'A'
        elif k >= 60:
            return 'B'
        elif k >= 55:
            return 'C'
        elif k >= 50:
            return 'D'
        else:
            return 'F'
    gv = np.vectorize(g)
    return np.apply_along_axis(g, 1, data)
c = grades(b)
print(c)
```

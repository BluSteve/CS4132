#### i) Based on the information in (h), plot a bar graph for each flat_type to compare the average resale price by town. You should use a loop where appropriate to generat the graph quickly. Note to label your graph clearly (axis and title).

```python
#your solution
flat_types = summary.columns.values
fig, ax = plt.subplots(len(flat_types),figsize=(20,80))
for t in range(len(flat_types)):
    ax[t].set_xlim([0,summary.max().max()*1.1]) # produces nice axis limits
    ax[t].barh(summary.index.values,summary[flat_types[t]])
    ax[t].set_title(flat_types[t][1])
```

#### b) You are to shortlist a question you wish to investigate (and not already covered in Question 1 or 2a) based on the data given.

i) State your question.

What is the relationship between the floor area of a flat and its resale price for new, moderate and old flats?

ii) Outline your analysis methodology and results (e.g. relevant table or visualization). Note to show each step clearly.

```python
#Now we will plot a scatterplot that has three hues, one for each flat_age_type
#The y-axis is price while the x-axis is floor area
#This visualizes how price changes with floor area for each age type.

sns.scatterplot(x='floor_area_sqm', y='resale_price', hue='flat_age_type',data=df)
```

```python
# Okay, I underestimated the amount of data points there are in this dataset.
# Let us categorize the floor area into bin of equal width, then
# find the average resale_price of each bin.
mi, ma = df['floor_area_sqm'].min(),df['floor_area_sqm'].max()
w = 20 # width of bin
start, end = (int(mi)//w)*w,(int(ma)//w)*w+w+1
print(mi,ma,(int(mi)//w)*w,(int(ma)//w)*w+w)
bins = [x for x in range(start,end,w)]
print (bins)
df['floor_area_binned'] = pd.cut(df['floor_area_sqm'],bins,labels=bins[:-1])
df
```

```python
# Now, let's group the price based on floor area type and age type
b = df.groupby(['floor_area_binned', 'flat_age_type'],as_index=False).resale_price.mean().round(2)
b.pivot(columns='flat_age_type', index='floor_area_binned')
```

```python
# Now, let's put it all together in a line graph
sns.lineplot(x='floor_area_binned',y='resale_price',hue='flat_age_type',data=b)
```

iii) Finally, answer the question based on the results from (ii)

In general, for all three flat_age_types, the larger the floor area, the higher the resale_price. Towards the far right, there are some outliers, perhaps mainly due to sparseness of data there. As flats get newer, the maximum floor area also decreases. For "New" flats, there are no flats with a floor area greater than 150, but for "Old" flats, the largest is around 300 sqm.

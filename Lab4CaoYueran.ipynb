### Q1

b) Find the mean score of each assessment component per student. Store the Student ID, Mean Assignment score, Mean Graded Score, Mean Test Score and Mean Exam score in one dataframe.
Print the dataframe.

```python
x = pd.DataFrame({k:v.iloc[1:,1:].mean(axis=1) for k,v in df.items()})
x.insert(0, 'Student ID', df['Assignment']['Student ID'])
x
```

c) Find the mean score of each assessment component. Store the result in a dataframe.

```python
#your solution
c = pd.DataFrame(x.mean())
c
```

### Q3

a) Read in the data from the following website: https://www.contextures.com/xlSampleData01.html
Store the data in a dataframe, with column names OrderDate etc.

```python
#your solution
import requests
url = requests.get('https://www.contextures.com/xlSampleData01.html')
q = pd.read_html(url.text)[0]
df1 = pd.DataFrame(q[1:])
df1.columns = q.loc[0]
df1
```

b) Find the total price for each item in a region.

```python
#your solution
df1.Total = df1.Total.astype(float)
df1.groupby(['Region', 'Item']).Total.sum()
```

c) Find the units per item sold by each Rep.

```python
#your solution
df1.Units = df1.Units.astype(int)
c = df1.groupby(['Rep', 'Item']).Units.sum()
c
```
